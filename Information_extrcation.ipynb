```python
%%writefile app.py
import streamlit as st
import csv
import re
#import pdf2doi
from PyPDF2 import PdfReader
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os
from langchain_google_genai import GoogleGenerativeAIEmbeddings
import google.generativeai as genai
from langchain.vectorstores import FAISS
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains.question_answering import load_qa_chain
from langchain.prompts import PromptTemplate
from dotenv import load_dotenv
import pandas as pd
import os

data = ["Paper Title","Problem", "Methodology","Dataset","Algorithm","Crops", "Future Direction","Research questions","Limitations","Main Findings","Results"]
# Set your Google API key
os.environ['GOOGLE_API_KEY'] = 'AIzaSyBA8p9Pf4St_zIHp7A19EZmu8bb4voIJwg'
responses = []
questions = [
    "Problem: Describe and explain the main problem solved in the paper.",
    "Methodology: Describe the research methods and experimental approaches in details employed in the study in details.",
    "Region: Describe the region where study is conducted?"
    "Dataset: Discuss the details regarding data used in paper for training and testing. Which spectral bands and features are used to solve the problem",
    "Algorithms: Enumerate the proposed algorithms, methods, techniques, models, or architectures utilized in the research, providing concise descriptions of their objectives and roles within the study.",
    "Crops: Identify the specific crops discussed in the study and any relevant agronomic details. Which spectral bands and features are used to solve the problem",
    "Future Research: Highlight suggestions for future research directions or unanswered questions the study proposes.",
    "Research Questions: Enumerate the primary research questions pr reserach objective the paper aims to address.",
    "Limitations: Outline any limitations or constraints acknowledged by the authors ",
    "Main Findings: Describe the conclusion of the paper",
    "Experimentation: Deatils on experimentation setup and their results and their accuracy?"
    "Results: Explain the resulsts and which metrics are used discussed in paper"
]
def get_pdf_text(pdf_docs):
    text=""
    for pdf in pdf_docs:
        pdf_reader= PdfReader(pdf)
        for page in pdf_reader.pages:
            text+= page.extract_text()
        doi_pattern = re.compile(r'10.\d{4,9}/[-._;()/:A-Z0-9]+', re.IGNORECASE)
        doi_matches = doi_pattern.findall(text)
    return  text,doi_matches



def get_text_chunks(text):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=1000)
    chunks = text_splitter.split_text(text)
    return chunks


def get_vector_store(text_chunks):
    embeddings = GoogleGenerativeAIEmbeddings(model = "models/embedding-001")
    vector_store = FAISS.from_texts(text_chunks, embedding=embeddings)
    vector_store.save_local("faiss_index")


def get_conversational_chain():

    prompt_template = """
    # You are research assistant and your duty is to answer questions thoroughly and accurately based on the provided context.
    # Ensure that all responses are concise, detailing all relevant information from the provided context.
    # If the information necessary for a response is not available within the provided context, clearly state that "the answer is not available in the context," and avoid giving incorrect information.
    # Responses should  encompass all necessary details to effectively address the inquiry.
    # Donot use the abbreviation.
    # Donot add anyother information other than context.
    Context:\n {context}?\n
    Question: \n{question}\n

    Answer:
    """

    model = ChatGoogleGenerativeAI(model="gemini-pro",
                             temperature=0)

    prompt = PromptTemplate(template = prompt_template, input_variables = ["context", "question"])
    chain = load_qa_chain(model, chain_type="stuff", prompt=prompt)

    return chain

def user_input(user_question):
    embeddings = GoogleGenerativeAIEmbeddings(model = "models/embedding-001")

    new_db = FAISS.load_local("faiss_index", embeddings,allow_dangerous_deserialization = True)
    docs = new_db.similarity_search(user_question,k=4)
    chain = get_conversational_chain()


    response = chain(
        {"input_documents":docs, "question": user_question}
        , return_only_outputs=True)
    return response["output_text"]
# Function to write values to a text file
def write_to_file(file_path, data):
      with open(file_path, 'w') as file:
          for value in data:
              file.write(str(value) + '\n')

# Function to read values from a text file
def read_from_file(file_path):
      data = []
      with open(file_path, 'r') as file:
          for line in file:
              data.append(line.strip())
      return data
def append_row_to_csv(row_data):
    try:
       file_path = '/content/drive/MyDrive/research_data.csv'
       with open(file_path, 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(row_data)
            st.write(row_data)
            st.write("Row appended successfully!")
    except Exception as e:
        st.write("An error occurred:", e)

def main():
    st.set_page_config("Article Data Extraction")
    st.header("Paper Extractor Pro")
    with st.sidebar:
        st.title("Menu:")
        pdf_docs = st.file_uploader("Upload your PDF Files and Click on the Submit & Process Button", accept_multiple_files=True)
        if st.button("Submit & Process"):
            with st.spinner("Processing..."):
                raw_text,doi = get_pdf_text(pdf_docs)
                text_chunks = get_text_chunks(raw_text)
                get_vector_store(text_chunks)
                st.success("Done")
                write_to_file("doi.txt", doi)




    #st.dataframe(df, use_container_width=True)
    user_question = st.text_input("Ask a Question about research paper")
    if user_question:
                doi = read_from_file("doi.txt")
                responses.append(doi[0])
                for question in questions:
                    response = user_input(question)
                    responses.append(response)
                df = pd.DataFrame()
                df=df.append(pd.Series(data),ignore_index=True)
                df=df.append(pd.Series(responses), ignore_index=True)
                # Calculate the maximum length of text across all columns for each row
                df['max_text_length'] = df.apply(lambda row: max([len(str(cell)) for cell in row]), axis=1)
                # Calculate the total height required for all rows
                total_height = sum(df['max_text_length'] // 30) + len(df) * 2
                st.dataframe(df,height=total_height)



                append_row_to_csv(responses)






if __name__ == "__main__":
    main()
```
